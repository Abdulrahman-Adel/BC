```python
categorical_cols = [col for col in df.columns if df[col].dtype == "object"]

categorical_corr = df[categorical_cols].apply(lambda x : pd.factorize(x)[0]).corr(method='pearson', min_periods=1)

sns.heatmap(categorical_corr,linewidth=0.5)
```
